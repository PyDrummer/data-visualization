# 500 Greatest Songs of All Time
### Rolling Stone's list of the 500 greatest songs of all time.
### I'm showing you the amount of hit songs per year.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
songs = pd.read_csv('./csv/top500.csv')
```

```python
def get_year(d):
    return int(d.split(',')[1])
```

```python
songs['year'] = songs['released'].apply(get_year)
songs = songs.sort_values(by='year', ascending=False)
year = songs['year'].value_counts(sort=False)
year_only = songs['year'].drop_duplicates().values
hits = year.values
```

```python
p = sns.relplot(x=year_only, y=hits, data=year, hue=hits, palette="flare", legend=False)
p.set(xlabel='Year', ylabel='Hits')
p.fig.suptitle("Number of hits songs by year", fontsize=20, y=1.05)
print()
```

## Insights

### As per rolling stone's top 500 hits it seems the most hits were during 1980s to the 1990s.

### Seems rolling stone's criteria of a 'hit' aren't being met going into the 200s
