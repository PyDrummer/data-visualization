# Top 10 highest producing famous artists of all time!
### We will take a look at which famous artist throughout the ages has created the most art.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
art = pd.read_csv('./csv/artists.csv')
top_ten = art.sort_values(by='paintings', ascending=False).head(10)
```

```python
plot = sns.barplot(x = 'paintings', y = 'name', data = top_ten, palette="icefire")
plot.set_title('Top 10 highest producing Artists of all time')
plot.set(xlabel='Paintings created', ylabel='Artist Name')
print()
```

## Insights

From our bar graph we can see that Vincent van Gogh is the definitive highest producing famous artist of all time.

Edgar Degas is in close second and Pablo Picasso

This data set shows only famous and recognized artists so there quite possibly could be a higher producing artist out in the world.
